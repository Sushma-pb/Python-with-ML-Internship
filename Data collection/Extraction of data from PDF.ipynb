# Extract the data from pdf files

```python
import PyPDF2 as pdf
```

```python
file = open('Sushma.pdf', 'rb')
```

```python
print(file)
```

```python
pdf_reader = pdf.PdfFileReader(file)
```

```python
file = pdf_reader
```

```python
help(pdf_reader)
```

```python
pdf_reader.getIsEncrypted() #to check whether our pdf encrypted or not, if it is false its not encrypted 
```

```python
pdf_reader.getNumPages() #it gives the total no. of pages 
```

```python
page1 =pdf_reader.getPage(0)# in .getPage(0) here (0):-is d index value 
```

```python
print(page1)
```

```python
d=page1.extractText() # here extrction of text from pdf
print(d)
```

```python
page2 =pdf_reader.getPage(1)
```

```python
page2
```

```python
print(page2)
```

```python
d1= page2.extractText()
```

```python
print(d1)
```

# # APPEND ,WRITE or Merge PDFs


```python
pdf_writer = pdf.PdfFileWriter()
```

```python
pdf_writer.addPage(page1)
pdf_writer.addPage(page2)
```

```python
output = open('Pages.pdf','wb')
pdf_writer.write(output)
output.close()
```
